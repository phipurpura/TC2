### Punto 1

Para lograr la plantilla pedida, hay qu tener todos los valores en función de $\omega$. Entonces: 

$f_{S1} = 17 kHz => \omega_{S1} = 2 * \pi * 17 kHz $

$f_{S2} = 34 kHz => \omega_{S2} = 2 * \pi * 34 kHz $

Una vez sacados estos valores, sacamos $\Omega$: 

$\omega_{S1 n} = \frac{\omega_{S1}}{\omega_{0}} = \frac{2 * \pi * 17 kHz}{2 * \pi * 22 kHz} = 0.7727 $

$\omega_{S2 n} = \frac{\omega_{S2}}{\omega_{0}} = \frac{2 * \pi * 34 kHz}{2 * \pi * 22 kHz} = 1.6363 $



Con esto ya se sabe la escala que lleva el eje de absisas. Queda saber que ocurre en el eje de ordenadas. Para eso se tiene que pasar de los valores del módulo de la transferencia a los de alfa. Entonces: 

$ |T(s)|_{dB} = - \alpha(s)_{dB} => $

$ |T(s)|_{dB max} = -0.5 dB =>  \alpha(s)_{dB max} = 0.5 dB $

$ |T(s)|_{dB min 1} = -16 dB =>  \alpha(s)_{dB min 1} = 16 dB$

$ |T(s)|_{dB min 2} = - 24 dB => \alpha(s)_{dB min 2} = 24 dB $

### Punto 2

Se tiene que epsilon es: 

$\epsilon^{2} = 10^{\frac{\alpha_{max}}{10}} - 1 = 0.122$

Implementando la transformación: 

$\Omega_{s1}  = Q * \frac{\omega_{S1 n}^{2}-1}{\omega_{S1 n}} = \frac {0.7727^{2}-1}{0.7727 } = -2.607  $

$\Omega_{s2}  = Q * \frac{\omega_{S2 n}^{2}-1}{\omega_{S2 n}} = \frac {1.6363^{2}-1}{1.6363 } = 5.125 $

Entonces, para el caso de un Chebyshev: 

$\alpha ( \omega_{s} ) = 10 . log ( 1 + \epsilon^{2} * cosh( n * arcosh( \omega_{s}))^{2} ) $

Iterando para el caso de $|\Omega_{s}|  = 2.607: $

$\alpha ( \omega_{s} ) = 10 . log ( 1 + 0.122 * cosh( 1 * arcosh( 2.607))^{2} ) = 2.622 $

$\alpha ( \omega_{s} ) = 10 . log ( 1 + 0.122 * cosh( 2 * arcosh( 2.607))^{2} ) = 13.0856 $

$\alpha ( \omega_{s} ) = 10 . log ( 1 + 0.122 * cosh( 3 * arcosh( 2.607))^{2} ) = 26.8672$

Iterando para el caso de $|\Omega_{s}|  = 5.125: $

$\alpha ( \omega_{s} ) = 10 . log ( 1 + 0.122 * cosh( 1 * arcosh( 5.125))^{2} ) = 6.2375 $

$\alpha ( \omega_{s} ) = 10 . log ( 1 + 0.122 * cosh( 2 * arcosh( 5.125))^{2} ) = 25.119 $


Tenemos entonces que para cumpla con la plantilla, se necesita un n = 3. Entonces la función queda: 

$|T(j \omega)| = \frac{1}{1+C_{n}^{2}(\omega) }$

Siendo:

$C_{n}(W) = 4 * \omega^{3} - 3*\omega $

$|T(j \omega)|^{2} = \frac{1}{1+ \epsilon^{2} * (4 * \omega^{3} - 3*\omega)^{2} } $

$|T(j \omega)|^{2} = \frac{1}{1+ 16 * \omega^{6} * \epsilon^{2} - 24*\omega^{4}* \epsilon^{2} + 9 * \omega^{2}* \epsilon^{2} } $

$|T(s)|^{2} = T(s) * T (-s)  = \frac{1}{-s^{6} * 16 * \epsilon^{2} - s^{4} * 24 * \epsilon^{2} - s^{2}*9*\epsilon^{2} + 1} = \frac{1}{s^{3} * a + s^{2} * b + s * c + d} * \frac{1}{-s^{3} * a + s^{2} * b - s * c + d}$ 


Entonces se tiene que: 

$ -a^{2} = -16 * \epsilon^{2} => a = 4 * \epsilon = 0.488 $

$ d = 1 $

$ b^{2} = 2*a*c + 24 * \epsilon^{2} => b = 3.2623 $

$ c^{2} = 2*b*d + 9 * \epsilon^{2} => c = 2.7609$

Por lo cual, se obtiene: 

$T(s) = \frac{1}{s^{3} * 0.488 + s^{2} * 3.2623 + s * 2.7609 + 1} = \frac{1}{0.488 * (s^{3} + s^{2} * \frac{3.2623}{0.488} + s * \frac{2.7609}{0.488} + \frac{1}{0.488})}  = \frac{2.049}{s^{3} + s^{2} * 6.685 + s * 5.6575 + 2.049} = \frac{2.049}{(s + 5.7653)(s^{2} + s * 0.9196 + 0.3553)}$

Con lo cual queda que: 

$\omega_{o1} = 5.7653 $

$\omega^{2}_{o2} = 0.3553 $

$\frac{Q_{2}}{\omega_{o2}} = 0.9196 => Q_{2} = 0.9196 * 0.5960 = 0.5481$

$ K = k_{1} * k_{2} * M =2.049 => k_{1} = 0.1734 , k_{2}= 2.8145 y M = 4.1984 $

Entonces: 

$T(s)_{PB} = 4.1984 * \frac{0.1734}{s + 5.7653} * \frac{2.8145}{s^{2} + s * 0.9196 + 0.3553}$

Con esto tenemos obtenida la transferencia del prototipo pasabajo que satisfaga el requerimiento del filtro pasabanda. Lo que falta es normalizarla, con lo cual quedaría: 

$T(s)_{PB} = M * \frac{1}{s + 1} * \frac{1}{s^{2} + s * \frac{1}{Q_{2}} + 1} = 4.1984 * \frac{1}{s + 1} * \frac{1}{s^{2} + s * \frac{1}{0.5481} + 1}$

### Punto 3

Para obtener la transferencia pasabanda normalizada se tiene que: 

$T(s) = T(Q * \frac{s^{2}+1}{s})_{PB} = M * \frac{1}{Q*(\frac{s^{2}+1}{s}) +1} * \frac{1}{Q^{2}*(\frac{s^{2}+1}{s})^{2}+Q*(\frac{s^{2}+1}{s})*\frac{1}{Q_{2}} +1} $

$T(s) = M * \frac {\frac{s}{Q}}{s^{2} + s * \frac{1}{Q} + 1} * \frac{ \frac{s^{2}}{Q^{2}} }{ s^{4}+s^{3}*\frac{1}{Q * Q_{2}} + s^{2} * (\frac{1}{Q^{2}}+1) + s * \frac{1}{Q * Q_{2}} + 1 } $
